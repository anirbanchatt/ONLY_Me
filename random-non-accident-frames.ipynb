## Generate random non accident frames from videos

```python
import cv2
import os
import PIL
import random
import shutil
```

```python
target_folder = 'dataset/frames/'
data_folder = 'dataset/RoadAccidents/'
destination_folder = 'dataset/randomNonAccidentFrames/'

list = os.listdir(data_folder)
```

```python
def extract_frames(target_folder, video_path, video_num):
    cap = cv2.VideoCapture(video_path)
    
    frame_count = 1
    while cap.isOpened():
        ret, frame = cap.read()
        
        if ret == False:
            break
        
        cv2.imwrite(target_folder + "vid" + str(video_num) + "-shot" + str(frame_count) + ".jpg", frame)
        frame_count += 1
        
        # cv2.imshow('Video', frame)
        if cv2.waitKey(1) == 13: # 13 is enter
            break

    cap.release()
    cv2.destroyAllWindows()
    
    return frame_count
```

```python
def generate_random_non_accident_frames(target_folder, destination_folder, frame_count, video_num):
    non_accident_frame_list = random.sample(range(1,frame_count), 12)
    print(non_accident_frame_list)
    for frame_num in non_accident_frame_list:
        frame_name = "vid" + str(video_num) + "-shot" + str(frame_num) + ".jpg"
        frame_path = target_folder + "/" + frame_name
        shutil.copy(frame_path, destination_folder)
    print('copied random non accident frames')
    
    l = os.listdir(target_folder)
    for f in l:
        os.remove(os.path.join(target_folder, f))
    print('deleted frames for vid', video_num)
```

```python
video_start_num = 120
video_end_num = len(list)
```

```python
for i in range(video_start_num, video_end_num):
    video_path = data_folder + list[i]
    video_num = int(list[i][list[i].find('s')+1:list[i].find('_')])
    frame_count = extract_frames(target_folder, video_path, video_num)
    generate_random_non_accident_frames(target_folder, destination_folder, frame_count, video_num)
    print('extracted ', list[i], 'frames')
```
